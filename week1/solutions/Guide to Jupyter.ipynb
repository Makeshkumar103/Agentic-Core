A Beginner’s Guide to the Command Line in Windows (using PowerShell)

This guide will help you get started with the command line on Windows using PowerShell. PowerShell conveniently supports Linux-like aliases such as ls, cp, mv, and rm, making it feel more familiar to those with a Unix/Linux background.

Part 1: Preliminaries
1.1 What is the Command Prompt (PowerShell) and Why Use It?
Command Prompt vs. PowerShell: In Windows, the term "Command Prompt" often refers to the older cmd.exe. However, PowerShell is a more modern and powerful shell that comes preinstalled on Windows 10 and later.
Why Use It?: The command line allows you to:
Navigate and manage files/directories efficiently.
Automate tasks via scripts.
Access developer tools (git, Python, etc.) more directly.
1.2 How to Open a PowerShell Prompt
Press Windows Key (or click the Start menu).
Type PowerShell.
Click Windows PowerShell or PowerShell (in Windows 11, it may be Windows Terminal with a PowerShell profile).

You’ll see a window with a prompt that typically looks like:

PS C:\Users\YourName>
1.3 Files and Directory Structure
Directories (Folders): Organizational units that can contain files or other directories (subdirectories).
Files: Individual items of data, e.g., documents, images, scripts.

On Windows, the typical path to your home directory is:

C:\Users\YourName
1.4 Printing the Current Directory

In PowerShell, you can see your current directory (the folder you are "in") with:

pwd

“pwd” stands for print working directory. Alternatively, you can use:

Get-Location

But pwd is the concise alias, more reminiscent of Linux.

Part 2: Common Commands in PowerShell
2.1 List Files: ls

To see a list of files and subdirectories in your current directory:

ls

In PowerShell, ls is actually an alias for Get-ChildItem, which shows items in the current location.

2.2 Move Between Directories: cd

To move into a directory, use:

cd name_of_directory

To move “up” one level, use:

cd ..

To return to your home directory, either type:

cd ~

or simply do:

cd C:\Users\YourName
2.3 Moving and Copying Files

Move (rename) a file:

mv old_file_name new_file_name

or to move a file to another directory:

mv file.txt path\to\other\directory\

Copy a file:

cp file.txt backup_file.txt

or to copy to a different directory:

cp file.txt C:\Users\YourName\somewhere\
2.4 The User’s Home Directory and a Typical “projects” Folder

Your home directory is:

C:\Users\YourName

It’s common to keep development or technical projects in a subfolder called projects. To create and enter that directory, you can do:

mkdir projects
cd projects
2.5 Creating and Deleting Directories/Files

Create a directory:

mkdir new_folder

Delete a file:

rm filename.txt

Delete a directory (and its contents):

rm -Recurse directory_name

Here, -Recurse tells PowerShell to remove all contents of the directory.

Warning: Deletions using the command line are often permanent (they don’t go to the Recycle Bin by default). Be careful.

Part 3: Hidden Files

On Unix-like systems, files that begin with a . (dot) are considered hidden. Windows treats hidden files differently at the system level, but PowerShell can still display them similarly.

3.1 Viewing Hidden Files

To list hidden files in PowerShell, use:

ls -Force

This shows all files, including hidden ones (Windows system files may also appear).

3.2 Creating a Hidden File (e.g., .env)

Create the file:

New-Item .env

or equivalently:

echo "MY_VARIABLE=123" > .env

Edit the file (using Notepad):

notepad .env

Type in what you need, save, and close.

See its contents:

cat .env

or

type .env

In PowerShell, cat is an alias for Get-Content.

Part 4: PowerShell vs. Anaconda Prompt
PowerShell is a general-purpose shell and scripting environment for Windows. It is not specifically tied to any Python or data science environment by default.
Anaconda Prompt looks similar but comes with Anaconda (a Python distribution for data science). When you open the Anaconda Prompt, it automatically:
Activates a conda environment (e.g., base).
Places Python and conda commands on your PATH, so you can directly run python or conda commands.

Both PowerShell and Anaconda Prompt allow you to run the same commands for file management (ls, cd, rm, etc.). The primary difference is that the Anaconda Prompt has the environment for data science and Python ready to go, whereas in PowerShell you might need to manually activate an environment (e.g., conda activate myenv).

Part 5: Exercises

Below are a few practice exercises to build your confidence with the command line. Try them out in PowerShell or in the Anaconda Prompt.

Check Your Current Location

Open PowerShell, type:

pwd

Confirm it shows your user directory.

Create and Navigate a Folder

Inside your home directory, create a folder called test_folder.

mkdir test_folder

Move into test_folder.

cd test_folder

Print your current directory to confirm.

pwd
Create Files and Subdirectories

Create an empty file called notes.txt.

New-Item notes.txt

Create a subdirectory called archive.

mkdir archive
Use ls to see your files and folders.
Move and Copy Files

Move notes.txt into archive.

mv notes.txt archive

Change into archive and copy notes.txt back up one directory (duplicate it).

cd archive
cp notes.txt ..
Remove Files and Directories

Remove the file notes.txt in your current directory.

rm notes.txt

Go back up to test_folder and remove archive entirely:

cd ..
rm -Recurse archive
Verify it’s gone with ls.
Hidden Files

Within test_folder, create a hidden file named .example:

New-Item .example

Check if it’s visible with ls. If not, use:

ls -Force

Remove .example:

rm .example
Working with .env Files

Create a .env file in your test_folder:

echo "SECRET_TOKEN=abc123" > .env
Use cat .env to verify the contents.
Optionally, edit it using notepad .env.
Bonus: Check the difference between PowerShell and Anaconda Prompt
Open the Anaconda Prompt.
Check the active conda environment (it should say (base) or similar in the prompt).
Compare by opening a normal PowerShell window and running python --version. If Python is not on the path, you’ll see a difference.
